```python
import re
import os
os.environ["OPENAI_API_KEY"] = "" # TODO

from openai import OpenAI
from typing import Dict, List

DEFAULT_PROMPT="""\
The algorithms are described as follows:
-------------------
{context}
-------------------
According to the algorithms above, transform the number {query} into a word. Return this word using the following format: [word].
"""

class kata_query_engine():
  def __init__(
      self,
      index: Dict,
      model: str = 'gpt-4-turbo',
  ) -> None:
    self.index = index
    self.model = model
    self._client = client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))
  def query(self, query: List, language: str):
    if language not in ['fr','bg']:
      raise ValueError("Language is not supported.")
    if all(isinstance(item, int) for item in query) == False:
      raise ValueError("Query needs to be numbers.")
    results = []
    for num in query:
      nodes = self.retrieve(num, language)
      context = ''
      for node in nodes:
        context += node['text']
      prompt = self.update_prompt(num, context)
      response = self.get_response(prompt)
      results.append(response)
    return results

  def get_response(self, prompt:str):
    chat_completion = self._client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": prompt,
            }
        ],
        model=self.model,
    )
    return self._reponse_parser(chat_completion)

  def _reponse_parser(self, chat_completion):
    match = re.search(r'\[(.*?)\]', chat_completion.choices[0].message.content)
    parsed_response = ''
    if match:
      parsed_response = match.group(1)

    start_letter_index = 0
    end_letter_index = 0
    for pos in range(len(parsed_response)):
      if parsed_response[pos].isalpha():
        start_letter_index = pos
        break
    for pos in range(len(parsed_response)):
      if parsed_response[-pos].isalpha():
        end_letter_index = len(parsed_response) - pos+1
        break
    return parsed_response[start_letter_index:end_letter_index].replace("-"," ")

  def update_prompt(self, num: int, context: str):
    prompt = DEFAULT_PROMPT
    prompt = prompt.replace("{context}",context)
    prompt = prompt.replace("{query}", str(num))
    return prompt

  def retrieve(self, num: int, language: str):
    nodes = []
    if num > 99:
      if language == "fr":
        nodes.append(self.index["node_100_fr"])
      elif language == "bg":
        nodes.append(self.index["node_100_bg"])

    decimal_num = num%100
    if decimal_num in list(range(17)):
      nodes.append(self.index["node_0_16"])
    else:
      if language == "fr":
        nodes.append(self.index["node_10_fr"])
        if decimal_num < 70:
          if decimal_num%10 == 1:
            nodes.append(self.index["node_end_1"])
          else:
            nodes.append(self.index["node_22_29"])
        else:
          nodes.append(self.index["node_70_fr"])
          if decimal_num%10 == 1:
            nodes.append(self.index["node_70_end_1_fr"])
      elif language == "bg":
        nodes.append(self.index["node_10_bg"])
        if decimal_num%10 == 1:
          nodes.append(self.index["node_end_1"])
        else:
          nodes.append(self.index["node_22_29"])
    return nodes
```
